## Importing the libraries

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

## Loading the dataset

```python
df = pd.read_csv("Exam data.csv")
```

```python
df
```

```python
df.isna().any()
```

```python
df.isnull().sum()
```

```python
df.hours = df.hours.fillna(df.hours.mean())
print(df)
```

## Segregate the dataset into X and Y

```python
X = df.iloc[:,:-1].values
X
```

```python
Y = df.iloc[:,-1].values
Y
```

## Training the model by Linear regression

```python
model = LinearRegression()
model.fit(X,Y)
```

## Predicting the Model

```python
a = [[8.67,23,6]]
PredictModel = model.predict(a)
print(PredictModel)
```
